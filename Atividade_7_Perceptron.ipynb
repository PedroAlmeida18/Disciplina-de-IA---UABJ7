```python

# Dados da questão:
# Estudou: 0/1
# Fez trabalho: 0/1
# Passou: 0/1

# Exemplo:
# Joãozinho: [1,0,0] → 0
# Huguinho:  [1,0,1] → 0
# Zezinho:   [1,1,0] → 1
# Luizinho:  [1,1,1] → 1
print("Atividade 7 : Perceptron  Aluno : Pedro Henrique de Almeida Santos e Cleyson Silva de Oliveira")

dados = [
    ([1,0,0], 0, "Joãozinho"),
    ([1,0,1], 0, "Huguinho"),
    ([1,1,0], 1, "Zezinho"),
    ([1,1,1], 1, "Luizinho")
]


alpha = 0.1


w = [0.0, 0.0, 0.0]

log = []
log.append("TREINAMENTO DO PERCEPTRON — MODELO IGUAL AO DA AULA\n")
log.append("Pesos iniciais: θ0=0, θ1=0, θ2=0\n\n")



def h(v):
    return 1 if v >= 0 else 0


for ciclo in range(1, 3):
    log.append(f"\n============================\nCICLO {ciclo}\n============================\n")

    for x, y, nome in dados:


        s = w[0]*x[0] + w[1]*x[1] + w[2]*x[2]
        out = h(s)
        erro = y - out

        log.append(f"\nEntrada: {nome}")
        log.append(f"x = {x}, y = {y}")
        log.append(f"s_out = h({w[0]}*{x[0]} + {w[1]}*{x[1]} + {w[2]}*{x[2]}) = h({s}) = {out}")
        log.append(f"Erro: E = {y} - {out} = {erro}")


        if erro != 0:
            old_w = w.copy()
            w[0] += alpha * erro * x[0]
            w[1] += alpha * erro * x[1]
            w[2] += alpha * erro * x[2]

            log.append("Atualização dos pesos:")
            log.append(f"θ0 = {old_w[0]} + {alpha} * {erro} * {x[0]} = {w[0]}")
            log.append(f"θ1 = {old_w[1]} + {alpha} * {erro} * {x[1]} = {w[1]}")
            log.append(f"θ2 = {old_w[2]} + {alpha} * {erro} * {x[2]} = {w[2]}")
        else:
            log.append("Sem atualização (E = 0)")

log.append("\nPesos finais:")
log.append
log.append(f"θ0 = {w[0]}, θ1 = {w[1]}, θ2 = {w[2]}\n")

with open("Resultados_Perceptron.txt", "w") as f:
    f.write("\n".join(log))

print("\n".join(log))
print("\nArquivo gerado: Resultados_Perceptron.txt")
```
